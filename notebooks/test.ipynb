```python
sub = gdf[gdf['statename'] == 'Andhra Pradesh']
```

```python
sub.geom_type
```

```python
lats = []; lons = []
for idx, row in sub.iterrows():
    if (row['geometry'].geom_type == 'MultiPolygon'):
        for poly in list(row['geometry']):
            lon_vals, lat_vals = poly.exterior.coords.xy
            lats.extend(list(lat_vals))
            lons.extend(list(lon_vals))
    else:
        lon_vals, lat_vals = row['geometry'].exterior.coords.xy
        lats.extend(list(lat_vals))
        lons.extend(list(lon_vals))
```
